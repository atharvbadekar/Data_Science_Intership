```python
import os
import openpyxl
import sys

# Create a new workbook
workbook = openpyxl.Workbook()
sheet = workbook.active

# Write the header row
sheet['A1'] = 'Fruit'
sheet['B1'] = 'Quantity'

# Write the data rows
sheet['A2'] = 'Apple'
sheet['B2'] = 10
sheet['A3'] = 'Banana'
sheet['B3'] = 5
sheet['A4'] = 'Apple'
sheet['B4'] = 15
sheet['A5'] = 'Orange'
sheet['B5'] = 20
sheet['A6'] = 'Banana'
sheet['B6'] = 10

# Save the workbook as original.xlsx
original_file = 'original.xlsx'
workbook.save(original_file)

# Close the workbook
workbook.close()

print("The original.xlsx file has been created.")

# Load the original spreadsheet
workbook = openpyxl.load_workbook(original_file)
sheet = workbook.active

# Create a dictionary to store the total quantity of each fruit
fruit_totals = {}

# Iterate over the rows in the spreadsheet
for row in sheet.iter_rows(min_row=2, values_only=True):
    fruit = row[0]
    quantity = row[1]

    # Update the total quantity for the fruit
    if fruit in fruit_totals:
        fruit_totals[fruit] += quantity
    else:
        fruit_totals[fruit] = quantity

# Create a new workbook for the results
result_workbook = openpyxl.Workbook()
result_sheet = result_workbook.active

# Write the header row
result_sheet['A1'] = 'Fruit'
result_sheet['B1'] = 'Total Quantity'

# Write the fruit and total quantity to the result spreadsheet
row_index = 2
for fruit, total_quantity in fruit_totals.items():
    result_sheet.cell(row=row_index, column=1).value = fruit
    result_sheet.cell(row=row_index, column=2).value = total_quantity
    row_index += 1

# Save the result workbook
result_file = 'result.xlsx'
result_workbook.save(result_file)

# Close the workbooks
workbook.close()
result_workbook.close()

print("Data has been processed and saved to", result_file)

# Open the original and result files to view the output
if sys.platform == 'win32':
    os.startfile(original_file)
    os.startfile(result_file)
elif sys.platform == 'darwin':  # macOS
    os.system('open ' + original_file)
    os.system('open ' + result_file)
elif sys.platform.startswith('linux'):  # Linux
    os.system('xdg-open ' + original_file)
    os.system('xdg-open ' + result_file)
else:
    print("Unable to open files automatically on this platform. Please manually open the files:", original_file, result_file)
```
